```python

from PIL import Image

def remove_corrupted_images(directory):
    files = [f for f in os.listdir(directory) if os.path.isfile(os.path.join(directory, f))]
    removed_count = 0

    for file in files:
        file_path = os.path.join(directory, file)
        try:
            with Image.open(file_path) as img:
                img.verify()  # Verifying the image integrity
        except (IOError, SyntaxError) as e:
            print(f"Removing corrupted image: {file_path}")
            os.remove(file_path)  # Removing the corrupted file
            removed_count += 1

    return removed_count

def count_files(directory):

    _, _, files = next(os.walk(directory))
    return len(files)

cat_images_directory = '/content/PetImages/Cat/'


initial_count = count_files(cat_images_directory)
print('Initial number of images:', initial_count)

# Removing corrupted images
removed_images_count = remove_corrupted_images(cat_images_directory)
print(f'Removed {removed_images_count} corrupted images.')

# Counting the files again after removal
final_count = count_files(cat_images_directory)
print('Final number of images:', final_count)
```

```python

image_directory = '/content/image_resized/'

# Initializing labels and filenames
filenames = os.listdir(image_directory)
labels = [1 if filename.startswith('dog') else 0 for filename in filenames]

# Splitting data
train_filenames, test_filenames, train_labels, test_labels = train_test_split(
    filenames, labels, test_size=0.2, random_state=42
)
```

```python
from torchvision import transforms
from torch.utils.data import Dataset, DataLoader


class ImageDataset(Dataset):
    def __init__(self, file_names, labels, directory, transform=None):
        self.file_names = file_names
        self.labels = labels
        self.directory = directory
        self.transform = transform

    def __len__(self):
        return len(self.file_names)

    def __getitem__(self, idx):
        img_path = os.path.join(self.directory, self.file_names[idx])
        image = Image.open(img_path)
        if self.transform:
            image = self.transform(image)
        return image, self.labels[idx]

# Defining transformations
train_transform = transforms.Compose([
    transforms.RandomHorizontalFlip(), transforms.RandomRotation(20),
    transforms.ToTensor(), transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])
test_transform = transforms.Compose([
    transforms.ToTensor(), transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# Creating DataLoaders
train_dataset = ImageDataset(train_filenames, train_labels, image_directory, transform=train_transform)
test_dataset = ImageDataset(test_filenames, test_labels, image_directory, transform=test_transform)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32)
```

```python
import torchvision.models as models
import torch.nn as nn
import torch

#  ResNet-18
resnet18 = models.resnet18(pretrained=True)
resnet18.fc = nn.Linear(resnet18.fc.in_features, 2)


device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
resnet18 = resnet18.to(device)
```

```python
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(resnet18.parameters(), lr=0.001)
```

```python
def train_model(model, criterion, optimizer, train_loader, test_loader, num_epochs=5):
    history = {'train_loss': [], 'train_acc': [], 'test_loss': [], 'test_acc': []}

    for epoch in range(num_epochs):
        model.train()
        train_loss, train_correct = 0, 0
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            train_loss += loss.item() * inputs.size(0)
            _, preds = torch.max(outputs, 1)
            train_correct += torch.sum(preds == labels.data)

        train_acc = 100. * train_correct.double() / len(train_loader.dataset)
        history['train_loss'].append(train_loss / len(train_loader.dataset))
        history['train_acc'].append(train_acc)

        model.eval()
        test_loss, test_correct = 0, 0
        with torch.no_grad():
            for inputs, labels in test_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                test_loss += loss.item() * inputs.size(0)
                _, preds = torch.max(outputs, 1)
                test_correct += torch.sum(preds == labels.data)

        test_acc = 100. * test_correct.double() / len(test_loader.dataset)
        history['test_loss'].append(test_loss / len(test_loader.dataset))
        history['test_acc'].append(test_acc)

        print(f'Epoch {epoch+1}/{num_epochs}: '
              f'Train Loss: {history["train_loss"][-1]:.4f}, '
              f'Train Acc: {history["train_acc"][-1]:.2f}%, '
              f'Test Loss: {history["test_loss"][-1]:.4f}, '
              f'Test Acc: {history["test_acc"][-1]:.2f}%')

    return history

history = train_model(resnet18, criterion, optimizer, train_loader, test_loader, num_epochs=10)
```

```python
import matplotlib.pyplot as plt


def plot_history(history):

    fig, (ax1) = plt.subplots(1, figsize=(5, 5))
    #fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10))
    fig.suptitle('Training and Validation Metrics')


    ax1.plot(history['train_loss'], label='Train Loss')
    ax1.plot(history['test_loss'], label='Test Loss')
    ax1.set_ylabel('Loss')
    ax1.set_xlabel('Epochs')
    ax1.legend()
    ax1.set_title('Loss over Epochs')


    plt.show()

plot_history(history)
```

```python

def plot_loss(history):
    plt.figure(figsize=(8, 5))
    plt.plot(history['train_loss'], label='Training Loss')
    plt.plot(history['test_loss'], label='Testing Loss')
    plt.title('Loss vs. Epochs')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.show()


plot_loss(history)
```

```python

from sklearn.metrics import roc_curve, auc, precision_recall_curve, average_precision_score

def evaluate_model(model, test_loader, device):
    model.eval()  # Setting the model to evaluation mode
    true_labels = []
    pred_scores = []

    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs = inputs.to(device)
            outputs = model(inputs)
            # Applying softmax to convert logits to probabilities
            probabilities = torch.nn.functional.softmax(outputs, dim=1)
            pred_scores.extend(probabilities[:, 1].cpu().numpy())
            true_labels.extend(labels.cpu().numpy())

    return true_labels, pred_scores

true_labels, pred_scores = evaluate_model(resnet18, test_loader, device)
```

```python
def plot_roc_curve(true_labels, pred_scores):
    fpr, tpr, _ = roc_curve(true_labels, pred_scores)
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic')
    plt.legend(loc="lower right")
    plt.show()

plot_roc_curve(true_labels, pred_scores)
```

```python
def plot_precision_recall_curve(true_labels, pred_scores):
    precision, recall, _ = precision_recall_curve(true_labels, pred_scores)
    average_precision = average_precision_score(true_labels, pred_scores)

    plt.figure(figsize=(8, 6))
    plt.plot(recall, precision, color='blue', lw=2, label=f'Precision-Recall curve (AP = {average_precision:.2f})')
    plt.fill_between(recall, precision, step='post', alpha=0.2, color='blue')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.ylim([0.0, 1.05])
    plt.xlim([0.0, 1.0])
    plt.title('Precision-Recall Curve')
    plt.legend(loc="lower left")
    plt.show()

plot_precision_recall_curve(true_labels, pred_scores)
```
