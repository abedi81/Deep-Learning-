```python
import os
from PIL import Image

def remove_corrupted_images(directory):
    files = [f for f in os.listdir(directory) if os.path.isfile(os.path.join(directory, f))]
    removed_count = 0

    for file in files:
        file_path = os.path.join(directory, file)
        try:
            with Image.open(file_path) as img:
                img.verify()  # Verifying the image integrity
        except (IOError, SyntaxError) as e:
            print(f"Removing corrupted image: {file_path}")
            os.remove(file_path)  # Removing the corrupted file
            removed_count += 1

    return removed_count

def count_files(directory):

    _, _, files = next(os.walk(directory))
    return len(files)


cat_images_directory = '/content/PetImages/Cat/'

# Counting the files initially
initial_count = count_files(cat_images_directory)
print('Initial number of images:', initial_count)

# Removing corrupted images and printing the count of removed images
removed_images_count = remove_corrupted_images(cat_images_directory)
print(f'Removed {removed_images_count} corrupted images.')

# Counting the files again after removal
final_count = count_files(cat_images_directory)
print('Final number of images:', final_count)
```

```python


def analyze_images(directory):
    files = os.listdir(directory)
    num_images = len(files)
    print(f"Number of images: {num_images}")

    for file in files:
        file_path = os.path.join(directory, file)
        try:
            with Image.open(file_path) as img:
                width, height = img.size
                mode = img.mode
                channels = {
                    '1': 1,  # Typically for black and white images
                    'L': 1,  # Greyscale
                    'P': 1,  # Palette (can vary)
                    'RGB': 3, # Red, Green, Blue
                    'RGBA': 4, # Red, Green, Blue, Alpha
                    'CMYK': 4  # Cyan, Magenta, Yellow, Black
                }.get(mode, 'Unknown number of channels')

                print(f"Image: {file} - Width: {width}, Height: {height}, Channels: {channels}")
        except IOError:
            print(f"Failed to open image {file_path}")

# Directory containing resized images
image_directory = '/content/image_resized/'

analyze_images(image_directory)
```

```python

import cv2

# Preparing labels and image data
image_directory = '/content/image_resized/'
files = [os.path.join(image_directory, f) for f in os.listdir(image_directory) if f.endswith(('png', 'jpg'))]
labels = [1 if 'dog' in file else 0 for file in files]
images = np.array([cv2.imread(file) for file in files])

# Converting list to numpy arrays
X = np.array(images)
Y = np.array(labels)
```

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Splitting data into training and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)

# Data augmentation setup
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    brightness_range=(0.8, 1.2),
    horizontal_flip=True,
    rescale=1./255,
    fill_mode='nearest'
)

# Applying the transformations to the training data and fitting the generator
datagen.fit(X_train)

#  X_test is scaled
X_test = X_test / 255.0
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GlobalAveragePooling2D, Dropout, Dense
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.regularizers import l2

# Using a pre-trained model as the base
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Partially unfreezing the top layers of the base model
for layer in base_model.layers[:-10]:
    layer.trainable = False

# Building the CNN
model = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    Dropout(0.5),
    Dense(256, activation='relu', kernel_regularizer=l2(0.001)),
    Dropout(0.3),  # Additional dropout layer
    Dense(1, activation='sigmoid')
])
```

```python


from tensorflow.keras.optimizers import Adam

# Compiling the model with an Adam optimizer
optimizer = Adam(learning_rate=0.0001)
model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])



```

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau

# Callbacks
early_stopping = EarlyStopping(monitor='val_loss', patience=10, verbose=1, mode='min')
model_checkpoint = ModelCheckpoint('best_model.h5', monitor='val_accuracy', save_best_only=True, mode='max', verbose=1)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.00001, verbose=1)

# Training the model
history = model.fit(
    datagen.flow(X_train, Y_train, batch_size=64),
    steps_per_epoch=len(X_train) // 64,
    epochs=10,
    validation_data=(X_test, Y_test),
    callbacks=[early_stopping, model_checkpoint, reduce_lr],
    verbose=1
)
```

```python
import matplotlib.pyplot as plt

# Plotting training & validation accuracy values
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Test')
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()

# Plot training & validation loss values
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Test')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()

plt.show()
```

```python

from sklearn.metrics import roc_curve, roc_auc_score, precision_recall_curve, average_precision_score

   #  Predicting probabilities for the test set
#Y_pred_probs = model.predict(X_test)
X_test_scaled = X_test / 255.0
Y_pred_probs = model.predict(X_test_scaled)
#  Computing the ROC curve and ROC AUC
fpr, tpr, thresholds = roc_curve(Y_test, Y_pred_probs)
roc_auc = roc_auc_score(Y_test, Y_pred_probs)

#  Computing the Precision-Recall curve and AUC
precision, recall, thresholds_pr = precision_recall_curve(Y_test, Y_pred_probs)
pr_auc = average_precision_score(Y_test, Y_pred_probs)

#  Plotting the ROC curve
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc="lower right")

#  Plotting the Precision-Recall curve
plt.subplot(1, 2, 2)
plt.plot(recall, precision, color='green', lw=2, label='Precision-Recall curve (area = %0.2f)' % pr_auc)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc="lower left")

plt.tight_layout()
plt.show()

```
