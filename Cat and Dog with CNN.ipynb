```python
import os
from PIL import Image

def remove_corrupted_images(directory):
    files = [f for f in os.listdir(directory) if os.path.isfile(os.path.join(directory, f))]
    removed_count = 0

    for file in files:
        file_path = os.path.join(directory, file)
        try:
            with Image.open(file_path) as img:
                img.verify()  # Verifying the image integrity
        except (IOError, SyntaxError) as e:
            print(f"Removing corrupted image: {file_path}")
            os.remove(file_path)  # Removing the corrupted file
            removed_count += 1

    return removed_count

def count_files(directory):

    _, _, files = next(os.walk(directory))
    return len(files)


cat_images_directory = '/content/PetImages/Cat/'


initial_count = count_files(cat_images_directory)
print('Initial number of images:', initial_count)


removed_images_count = remove_corrupted_images(cat_images_directory)
print(f'Removed {removed_images_count} corrupted images.')

final_count = count_files(cat_images_directory)
print('Final number of images:', final_count)
```

```python


def analyze_images(directory):
    files = os.listdir(directory)
    num_images = len(files)
    print(f"Number of images: {num_images}")

    for file in files:
        file_path = os.path.join(directory, file)
        try:
            with Image.open(file_path) as img:
                width, height = img.size
                mode = img.mode
                channels = {
                    '1': 1,  # Typically for black and white images
                    'L': 1,  # Greyscale
                    'P': 1,  # Palette (can vary)
                    'RGB': 3, # Red, Green, Blue
                    'RGBA': 4, # Red, Green, Blue, Alpha
                    'CMYK': 4  # Cyan, Magenta, Yellow, Black
                }.get(mode, 'Unknown number of channels')

                print(f"Image: {file} - Width: {width}, Height: {height}, Channels: {channels}")
        except IOError:
            print(f"Failed to open image {file_path}")


image_directory = '/content/image_resized/'

analyze_images(image_directory)
```

```python

image_directory = '/content/image_resized/'
files = [os.path.join(image_directory, f) for f in os.listdir(image_directory) if f.endswith(('png', 'jpg'))]
labels = [1 if 'dog' in file else 0 for file in files]
images = np.array([cv2.imread(file) for file in files])

# Convert list to numpy arrays
X = np.array(images)
Y = np.array(labels)
```

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Splitting data into training and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)

# Data augmentation setup
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    rescale=1./255,  # Apply scaling here
    fill_mode='nearest'
)

# Applying the transformations to the training data and fit the generator
datagen.fit(X_train)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization, GlobalAveragePooling2D
from tensorflow.keras.regularizers import l2

# Building the enhanced CNN
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3), padding='same'),
    BatchNormalization(),
    MaxPooling2D(2, 2),
    Conv2D(64, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D(2, 2),
    Conv2D(128, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D(2, 2),
    Conv2D(256, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D(2, 2),
    GlobalAveragePooling2D(),
    Dropout(0.5),
    Dense(256, activation='relu', kernel_regularizer=l2(0.01)),
    Dense(1, activation='sigmoid')
])
```

```python

from tensorflow.keras.optimizers import Adam

# Compiling the model with an Adam optimizer and learning rate decay
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer,
              loss='binary_crossentropy',
              metrics=['accuracy'])

```

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# Preparing callbacks for early stopping and best model saving
early_stopping = EarlyStopping(monitor='val_loss', patience=10, verbose=1, mode='min')
model_checkpoint = ModelCheckpoint('best_model.h5', monitor='val_accuracy', save_best_only=True, mode='max', verbose=1)

# Applying the existing datagen to fit the model
history = model.fit(
    datagen.flow(X_train, Y_train, batch_size=64),
    steps_per_epoch=len(X_train) // 64,
    epochs=15,
    validation_data=(X_test / 255, Y_test),
    callbacks=[early_stopping, model_checkpoint],
    verbose=1
)

```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Test')
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()


plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Test')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()

plt.show()
```

```python

probabilities = model.predict(X_test / 255).ravel()
```

```python
from sklearn.metrics import roc_curve, auc, precision_recall_curve, average_precision_score

# Computing ROC curve and ROC area
fpr, tpr, thresholds = roc_curve(Y_test, probabilities)
roc_auc = auc(fpr, tpr)

# Computing Precision-Recall curve and average precision
precision, recall, _ = precision_recall_curve(Y_test, probabilities)
average_precision = average_precision_score(Y_test, probabilities)
```

```python
import matplotlib.pyplot as plt

#  ROC curve
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")

#  Precision-Recall curve
plt.subplot(1, 2, 2)
plt.plot(recall, precision, color='blue', lw=2, label='Precision-Recall curve (AP = %0.2f)' % average_precision)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc="lower left")

plt.show()
```
